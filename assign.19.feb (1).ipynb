Q3.app route
App routing in Flask is the process of establishing the URL routes and their corresponding view functions for a Flask application. This is usually done 
using the @app.route decorator in the Flask application code

We use app routing in Flask to make it easy to create web applications with multiple pages or endpoints. By defining routes and associating them with view functions, we can create a flexible and modular application that can handle different types of requests and respond with different types of data.

```python
Q5.url_for() is used in flaskfor url building

from flask import Flask, url_for  #first flsask and url_for statements are imported
app = Flask(__name__)
@app.route('/')
def index():                                            #The index() function returns a simple "Hello, World!" message when the user navigates to the root URL (/)
    return 'Hello, World!'                              #The user_profile() function takes a username parameter from the URL and returns a personalized greeting
@app.route('/user/<username>')
def user_profile(username):
    return 'Welcome, {}!'.format(username)
if __name__ == '__main__':
    with app.test_request_context():                   #We use the test_request_context() function to create a test request context for our application. 
                                                       #This is necessary because Flask needs a request context in order to generate URLs.
        print(url_for('index'))                        #We then call url_for() twice, passing the name of each route as the first argument and any necessary parameter values as additional keyword arguments. The function returns a URL for each route, which we print to the console.

        print(url_for('user_profile', username='john'))
    app.run()
```
